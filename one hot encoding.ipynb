```python
import pandas as pd
```

```python
df=pd.read_csv('car-price.csv')
```

```python
df
```

```python
dummies=pd.get_dummies(df['Car Model'])
```

```python
dummies
```

```python
merged = pd.concat([df,dummies], axis='columns')
```

```python
merged
```

```python
final = merged.drop(['Car Model'], axis='columns')
```

```python
final
```

```python
y=df['sell price']
```

```python
y
```

```python
x=final.drop(['sell price'],axis=1)
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[45,7,1,0,0,0]])
```
